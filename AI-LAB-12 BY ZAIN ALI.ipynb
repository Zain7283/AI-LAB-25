```python

from flask import Flask, render_template, request
import pandas as pd
import pickle

app = Flask(__name__)

model = pickle.load(open("model.pkl", "rb"))

@app.route("/")
def home():
    return render_template("index.html")

@app.route("/predict", methods=["POST"])
def predict():
    val1 = float(request.form["v1"])
    val2 = float(request.form["v2"])
    val3 = float(request.form["v3"])
    val4 = float(request.form["v4"])

    df = pd.DataFrame([[val1, val2, val3, val4]])
    output = model.predict(df)[0]

    return render_template("index.html", result=output)

if __name__ == "__main__":
    app.run(debug=True)
```
